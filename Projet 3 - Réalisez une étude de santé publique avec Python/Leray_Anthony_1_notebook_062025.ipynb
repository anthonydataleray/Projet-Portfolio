```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017

#Je filtre les deux tables concernées pour l'année 2017
sous_nutrition_2017 = sous_nutrition[sous_nutrition['Année'] == '2016-2018']
population_2017 = population[population['Année'] == 2017]

#Je fais la fusion
pop_sous_nutri = pd.merge(sous_nutrition_2017, population_2017, on='Zone', how='inner')
```

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>
</div>

```python
#Combien mange en moyenne un être humain ? 

#Selon l'Oxfam qui relaie des chiffres des Etats-Unis, on peut retenir le chiffre de 2100 Kcal/jour par personne comme un seuil des besoins alimentaires essentiels.
#Source => https://www.oxfamfrance.org/agriculture-et-securite-alimentaire/faim-dans-le-monde-causes-consequences-solutions/#_ftnref3
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
population_2017 = population[population['Année'] == 2017]
dispo_alim_pop_merge = pd.merge(dispo_alimentaire, population_2017[['Zone', 'Population']], on='Zone', how='inner')
```

```python
#Affichage du nouveau dataframe  
dispo_alim_pop_merge.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
dispo_alim_pop_agrege = dispo_alim_pop_merge.pivot_table(
    index='Zone',
    aggfunc={
        'Disponibilité alimentaire (Kcal/personne/jour)': 'sum',
        'Population': 'first'
    }
)
dispo_alim_pop_agrege=dispo_alim_pop_agrege.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)': 'Disponibilité alimentaire total (Kcal/personne/jour)'})
dispo_alim_pop_agrege.head()
```

```python
#Calcul du nombre d'humains pouvant être nourris

#Rajout des deux colonnes 'Nombre de personnes pouvant être nourris' et 'Part de la population'
dispo_alim_pop_agrege['Nombre de personnes pouvant être nourris']=dispo_alim_pop_agrege['Disponibilité alimentaire total (Kcal/personne/jour)']*dispo_alim_pop_agrege['Population'] / 2100
dispo_alim_pop_agrege['Nombre de personnes pouvant être nourris']=dispo_alim_pop_agrege['Nombre de personnes pouvant être nourris'].astype(int)
dispo_alim_pop_agrege['Part de la population']=round(dispo_alim_pop_agrege['Nombre de personnes pouvant être nourris']/dispo_alim_pop_agrege['Population'] * 100,1)
dispo_alim_pop_agrege.head(5)

#Calcul du nombre de personnes à l'échelle mondiale
total_pop_mondiale=dispo_alim_pop_agrege['Population'].sum()
total_capacité_personnes_nourri = dispo_alim_pop_agrege['Nombre de personnes pouvant être nourris'].sum()
part_mondial_pers_nutri = round(total_capacité_personnes_nourri/total_pop_mondiale*100,1)

print(f"Au regard des capacités de production de nourriture, le nombre d'humains pouvant être nourris est de {total_capacité_personnes_nourri}, soit {part_mondial_pers_nutri} % de la population mondiale.")
```

```python
#Transfert des données avec les végétaux dans un nouveau dataframe
population_2017 = population[population['Année'] == 2017]
dispo_alimentaire_vegetaux=dispo_alimentaire[dispo_alimentaire['Origine'] == 'vegetale']
population_2017['Population'].astype(int)

dispo_alimentaire_vegetaux = dispo_alimentaire_vegetaux.drop(columns=[
    "Aliments pour animaux", 
    "Autres Utilisations", 
    "Disponibilité alimentaire en quantité (kg/personne/an)",
    "Disponibilité de matière grasse en quantité (g/personne/jour)",
    "Disponibilité de protéines en quantité (g/personne/jour)",
    "Disponibilité intérieure", "Exportations - Quantité", 
    "Importations - Quantité", 
    "Nourriture", 
    "Pertes",
    "Production",
    "Semences",
    "Traitement",
    "Variation de stock"])

pop_dispo_vegetaux_merge = pd.merge(dispo_alimentaire_vegetaux, population_2017, on = 'Zone', how='left')

display(pop_dispo_vegetaux_merge)
```

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.4 - Utilisation de la disponibilité intérieure</h3>
</div>

```python
#Calcul de la disponibilité totale

dispo_alim_mondiale = dispo_alimentaire.pivot_table(
    index ='Zone',
    values = 'Disponibilité intérieure',
    aggfunc ='sum'
)
display(dispo_alim_mondiale)
print(f"Le total de la disponibilité intérieure mondiale est de {dispo_alimentaire['Disponibilité intérieure'].sum():,} tonnes.")
```

```python
# Liste des colonnes qu'on veut analyser
colonnes_a_ajouter = ['Nourriture', 'Pertes', 'Semences', 'Traitement', 'Aliments pour animaux', 'Autres Utilisations' ]


# Étape 2 : pour chaque colonne dans notre liste (ex: 'Nourriture', 'Pertes')
for col in colonnes_a_ajouter:
    # On crée une table temporaire avec la somme de cette colonne par pays
    temp = dispo_alimentaire.pivot_table(
        index='Zone',
        values=col,
        aggfunc='sum'
    )

    # On crée une nouvelle colonne dans notre tableau principal avec le suffixe "_part"
    # Elle contient la part de cette catégorie (Nourriture, Pertes...) dans la dispo intérieure
    dispo_alim_mondiale[col + '_part'] = round(temp[col] / dispo_alim_mondiale['Disponibilité intérieure']*100,1)

display(dispo_alim_mondiale)
```

```python
#Calcul de la moyenne au niveau mondial
display(dispo_alim_mondiale['Nourriture_part'].mean())
display(dispo_alim_mondiale['Aliments pour animaux_part'].mean())
display(dispo_alim_mondiale['Pertes_part'].mean())
display(dispo_alim_mondiale['Semences_part'].mean())
display(dispo_alim_mondiale['Traitement_part'].mean())
display(dispo_alim_mondiale['Autres Utilisations_part'].mean())
```

```python
#Création d'une liste avec toutes les variables
df_cereales = dispo_alimentaire.loc[(dispo_alimentaire["Produit"] == "Millet") | (dispo_alimentaire["Produit"] == "Orge") | (dispo_alimentaire["Produit"] == "Seigle")|
                            (dispo_alimentaire["Produit"] == "Céréales, Autres")| (dispo_alimentaire["Produit"] == "Blé")| (dispo_alimentaire["Produit"] == "Maïs")|
                            (dispo_alimentaire["Produit"] == "Avoine") | (dispo_alimentaire["Produit"] == "Riz (Eq Blanchi)")| (dispo_alimentaire["Produit"] == "Sorgho")]
```

```python
#Création d'un dataframe avec les informations uniquement pour ces céréales
display(df_cereales)
```

```python
#Affichage de la proportion d'alimentation animale
df_cereales_animal_pivot = df_cereales.pivot_table(index='Zone', values=['Disponibilité intérieure','Aliments pour animaux'], aggfunc='sum')

# Rajour d'une colonne indiquant la part par pays
df_cereales_animal_pivot['Part alimentation animale (%)'] = round(
    (df_cereales_animal_pivot['Aliments pour animaux'] / df_cereales_animal_pivot['Disponibilité intérieure']) * 100, 1)
display(df_cereales_animal_pivot)

# Calcul de la part mondial
part_mondial_animal = df_cereales_animal_pivot['Aliments pour animaux'].sum()/df_cereales_animal_pivot['Disponibilité intérieure'].sum()*100
print(f"La proportion des céréales consacrée à l'alimnentation animale au niveau mondial est de {round(part_mondial_animal,1)} %.")
```

```python
#Affichage de la proportion d'alimentation humaine (correction consigne, il y avait deux fois alimentation animale)
df_cereales_humain_pivot = df_cereales.pivot_table(index='Zone', values=['Disponibilité intérieure','Nourriture'], aggfunc='sum')

# Rajour d'une colonne indiquant la part par pays
df_cereales_humain_pivot['Part alimetation humaine (%)'] = round(
    (df_cereales_humain_pivot['Nourriture'] / df_cereales_humain_pivot['Disponibilité intérieure']) * 100, 1)
display(df_cereales_humain_pivot)

# Calcul de la part mondial
part_mondial_humain = df_cereales_humain_pivot['Nourriture'].sum()/df_cereales_humain_pivot['Disponibilité intérieure'].sum()*100

print(f"La proportion des céréales consacrée à l'alimentation humaine au niveau mondial est de {round(part_mondial_humain,1)} %.")
```

```python
#Création d'un dataframe avec la zone, l'année et l'aide alimentaire puis groupby sur zone et année 

aide_alim_zone_année = aide_alimentaire.groupby(['Zone', 'Année']).sum()
aide_alim_zone_année = aide_alim_zone_année.drop(columns=['Produit'])
aide_alim_zone_année.head(10)
```
