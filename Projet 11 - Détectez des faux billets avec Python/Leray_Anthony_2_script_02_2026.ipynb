# 1 Import des librairies et du fichier à tester

```python
# ==========================================
# BLOC 1 : CONFIGURATION
# ==========================================
import pandas as pd
import joblib
from google.colab import drive

drive.mount('/content/drive/')
path = '/content/drive/MyDrive/Data Analyst/Module 12/'

# --- LA SEULE LIGNE À MODIFIER DEMAIN ---
nom_du_fichier = 'billets_complets_clean.csv'
# -----------------------------------------

path_csv = path + nom_du_fichier
```

# 2 Application de détection

```python
# ==========================================
# BLOC 2 : LE MOTEUR (LA FONCTION)
# ==========================================
def load_and_predict(input_file):
    try:
        model = joblib.load(path + 'best_model.joblib')
        scaler = joblib.load(path + 'scaler.joblib')

        data_to_test = pd.read_csv(input_file, sep=';') # Vérifier le sep demain
        features = ['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']
        X = data_to_test[features]

        X_scaled = scaler.transform(X)
        predictions = model.predict(X_scaled)
        probabilities = model.predict_proba(X_scaled)

        data_to_test['Prediction'] = ["VRAI" if p == 1 else "FAUX" for p in predictions]
        data_to_test['Probabilité'] = [f"{round(max(prob)*100, 2)}%" for prob in probabilities]

        return data_to_test
    except Exception as e:
        print(f"❌ Erreur : {e}")
        return None

# ==========================================
# BLOC 3 : EXÉCUTION ET RÉSULTATS
# ==========================================
df_resultats = load_and_predict(path_csv)

if df_resultats is not None:
    print(f"\n--- ANALYSE DU FICHIER : {nom_du_fichier} ---")
    print(df_resultats[['Prediction', 'Probabilité']])
    print("\n--- RÉPARTITION FINALE ---")
    print(df_resultats['Prediction'].value_counts())
```
